```python
import pandas as pd
import json
import requests
import os
import csv
import numpy as np
```

```python
newCars_df= pd.read_csv(r"Vehicle.csv", encoding='cp1252')
newCars_df
```

```python
newCars_df = newCars_df.filter(['ST_CASE', 'VEH_NO','BODY_TYPNAME','SPEEDRELNAME','DR_SF1NAME','P_CRASH2NAME','HIT_RUNNAME'], axis=1)
newCars_df
```

```python
newCars = newCars_df[~newCars_df.isin(['NOT APPLICABLE', 'Not Reported', 'Reported as Unknown', 'Other', 'None', 'NaN']).any(axis=1)]
newCars
```

```python
vehicle_df = newCars.rename(columns={"ST_CASE": "Case_Number", "VEH_NO": "Vehicle No.",
                                        "BODY_TYPNAME": "Body Type", "SPEEDRELNAME": "Speed Exceed", 
                                        "DR_SF1NAME": "Driver Safety", "P_CRASH2NAME": "Pedestrian Crash", 
                                        "HIT_RUNNAME": "Hit/Run"})
vehicle_df.head()
```

```python
vehicle_df.dtypes
```

```python
# Writing Accidents final dataframe to a csv
vehicle_df.to_csv(r"Output/vehicles_final.csv", index=False, header=True)
```
